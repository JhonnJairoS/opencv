```
import cv2
from matplotlib import pyplot as plt
image2=cv2.imread('casa.jpg')
plt.imshow(image2)
plt.show()
```

```
cv2.imwrite('casa.jpg', image2)
image2=cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
plt.show()

```

```
img_g = cv2.GaussianBlur(image2,(15,15),7) #se utiliza un kernel Gausiano
plt.imshow(img_g[...,::-1])
plt.show()
```

```
img_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
plt.imshow(img_gray[...,:],cmap='gray')
plt.show()
```

```
img_canny = cv2.Canny(image2,180,200)
plt.imshow(img_canny[...,::-1],cmap = 'gray')
plt.show()
```
